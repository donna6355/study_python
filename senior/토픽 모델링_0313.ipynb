```python
import gensim
from gensim import corpora, models
from gensim.corpora import Dictionary
```

### df 내부의 분석할 클러스터 정의
- 분석할 클러스터에 맞게 df를 수정하세요

```python
## 분석할 클러스터에 부합하게 df 수정
df_cluster2 = df[df["cluster"] == 2]


all_documents = list(df_cluster2["token"])
dictionary = Dictionary(all_documents)
dictionary.token2id
```

```python
# 확인용
corpus = []
for i in all_documents:
  corpus.append(dictionary.doc2bow(i))
corpus[0]
```

### 코히어런스와 퍼플렉시티 연산


*   perplexity, 낮을수록 happy
*   coherence, 높을수록 happy!


```python
from tqdm.auto import tqdm
import time
```

```python
from gensim.models import CoherenceModel
import matplotlib.pyplot as plt
import numpy as np

```

```python
perplexity_values = []

for i in tqdm(range(2, 11)) :
  ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=i, id2word = dictionary, random_state = 42)
  perplexity_v = ldamodel.log_perplexity(corpus)
  perplexity_values.append(perplexity_v)

x = range(2,11)
plt.plot(x, perplexity_values)
plt.xlabel('Number of topics')
plt.ylabel('Perplexity score')

```

```python
coherence_values = []
for i in tqdm(range(2,11)):
    ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=i, id2word = dictionary, random_state = 42)
    coherence_model = CoherenceModel(model=ldamodel, texts=all_documents, dictionary=dictionary, topn=5)
    coherence_score = coherence_model.get_coherence()
    coherence_values.append(coherence_score)

x = range(2,11)
plt.plot(x, coherence_values)
plt.xlabel('Number of topics')
plt.ylabel('Coherencce score')
```

### LDA 모델 구축
- num_topics 반드시 수정

```python
# num_topics 반드시 수정
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = 3, id2word = dictionary, random_state = 42)
```

```python
ldamodel.print_topics(num_words = 10)
```

```python
ldamodel.get_document_topics(corpus[1])
```

```python
align = []

for i in ldamodel.get_document_topics(corpus):
  label = []
  value = []
  for w in i:
    label.append(w[0])
    value.append(w[1])

  align.append(label[np.argmax(value)])

df_cluster2["topic"] = align
df_cluster2
```
