```python
import pandas as pd

df = pd.read_csv('blogs.csv')
df
```

```python
# df = df.rename(columns={'content': 'contents'})
# df
```

```python
df = df.dropna(subset=['contents'])
df.info()
df.head()
```

```python
del df['Unnamed: 0']
df
```

```python
import re

remove = re.compile(r'[^가-힣]')
df['contents'] = df['contents'].apply(lambda x : remove.sub(" ",x))
df['contents'] = df['contents'].apply(lambda x : x.replace("  "," "))
df['contents'] = df['contents'].apply(lambda x : x.replace("\n",""))
df
```

```python
#가입인사 제거, 가입, 반갑
df = df[~df['contents'].str.contains(r'반갑|가입', na=False, regex=True)]
df
```

```python

df['contents'] = df['contents'].apply(lambda x : x.strip())
df['len'] = df['contents'].apply(lambda x : len(x))
df = df.loc[df['len']>=20]
del df['len']
df = df.reset_index(drop=True)
df
```

```python
# ## ===== konlpy 설치 ===== ##
# from google.colab import drive
# drive.mount('/content/drive')

# # 코랩 konlpy 실행
# !curl -s https://raw.githubusercontent.com/teddylee777/machine-learning/master/99-Misc/01-Colab/mecab-colab.sh | bash

# # Mecab 설치 후 Google Drive에 복사
# !cp -r /usr/local/lib/mecab /content/drive/MyDrive/mecab
# !cp -r /usr/local/etc/mecabrc /content/drive/MyDrive/mecab
# ## ===== konlpy 설치 ===== ##

from konlpy.tag import Okt
from tqdm import tqdm
tqdm.pandas()

okt = Okt()
def extract_pos(text, pos):
  left = []
  for t,c in okt.pos(text, stem = True, norm = True):
    if c in pos:
      left.append(t)
  return left

df["token"] = df["contents"].progress_apply(lambda x : extract_pos(x,['Noun','Adjective','Verb']))
df
```

```python
# !pip install kiwipiepy
from kiwipiepy.utils import Stopwords

stopwords = Stopwords()
sw = set([i[0] for i in stopwords.stopwords])
sw
```

```python
sw.add('이다')
sw.add('하다')
sw.add('않다')
sw.add('그렇다')
sw.add('되다')
sw.add('있다')
sw.add('없다')
sw.add('싶다')
```

```python

cleaned_token = []
for i in df["token"]:
  cleaned_token.append([ w for w in i if w not in sw])
df["token"] = cleaned_token
df
```
