```python
#í¬ìŠ¤íŠ¸ ì œëª© ëª…ì‚¬ ì¶”ì¶œ
from konlpy.tag import Kkma

# Kkma ê°ì²´ ìƒì„±
kkma = Kkma()

# ëª…ì‚¬ ì¶”ì¶œ í•¨ìˆ˜ ì •ì˜
def extract_nouns(text):
    nouns = kkma.nouns(text)
    return nouns

df['title_ëª…ì‚¬'] = df['ì œëª©'].apply(extract_nouns)
print(df['title_ëª…ì‚¬'])
```

```python
# í•œ ê¸€ì ë¬¸ìì—´ì„ ì œê±°í•˜ëŠ” í•¨ìˆ˜
def remove_single_char(strings):
    return [string for string in strings if len(string) > 1]

df['title_ëª…ì‚¬'] = df['title_ëª…ì‚¬'].apply(remove_single_char)
df
```

```python
#ë¹ˆë„ ë¶„ì„ í›„ ë‹¨ì–´ ì •ì œ
from collections import Counter

# title ì—´ì˜ ë¬¸ìì—´ ë¦¬ìŠ¤íŠ¸ë¥¼ í•©ì¹˜ê¸°
all_keywords_title = [keyword for sublist in df['title_ëª…ì‚¬'] for keyword in sublist]

# ë¹ˆë„ ê³„ì‚°
title_counts = Counter(all_keywords_title)

# ë¹ˆë„ê°€ ë†’ì€ ìƒìœ„ 50ê°œ ì¶”ì¶œ
title_top_50_keywords = title_counts.most_common(50)

top_keywords = []
# ì¶œë ¥
print("ìƒìœ„ 50ê°œ ë¹ˆë„ê°€ ë†’ì€ í‚¤ì›Œë“œ(ëª…ì‚¬):")
for keyword, count in title_top_50_keywords:
  top_keywords.append({keyword:count})
  print(f"{keyword}: {count}íšŒ")
```

```python
keywords_df = pd.DataFrame([
    (list(d.keys())[0], list(d.values())[0])
    for d in top_keywords
], columns=["ë‹¨ì–´", "ë¹ˆë„ìˆ˜"])
keywords_df
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a dictionary of word frequencies
word_freq = dict(zip(keywords_df['ë‹¨ì–´'], keywords_df['ë¹ˆë„ìˆ˜']))

# Set font path for Korean characters
font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf' # Adjust if necessary

# Create a WordCloud object
wordcloud = WordCloud(
    font_path=font_path,
    width=800,
    height=400,
    background_color='white',
    colormap="viridis"
).generate_from_frequencies(word_freq)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('', fontsize=16)
plt.show()
```
