```python
import pandas as pd
import requests


KOSPI_URL='https://api.stock.naver.com/chart/domestic/index/KOSPI?periodType=dayCandle'
res = requests.get(KOSPI_URL)
res
```

res.json()

```python
datas = res.json()['priceInfos']
datas
```

```python
kospi_df = pd.DataFrame(datas)
kospi_df.head()
```

```python
#KOSDAQ
KOSDAQ = 'https://api.stock.naver.com/chart/domestic/index/KOSDAQ?periodType=dayCandle'

kosdaq = requests.get(KOSDAQ)
daq_data = kosdaq.json()['priceInfos']
kosdaq_df = pd.DataFrame(daq_data)
kosdaq_df.head()
```

```python
#USD Exchange Rate
USD = 'https://m.stock.naver.com/front-api/chart/pricesByPeriod?reutersCode=FX_USDKRW&category=exchange&chartInfoType=marketindex&scriptChartType=areaYear'
usd = requests.get(USD)
usd
usd_data = usd.json()['result']['priceInfos']
usd_df = pd.DataFrame(usd_data)
usd_df.head()
```

```python
print('KOSPI')
display(kospi_df)
print('KOSDAQ')
display(kosdaq_df)
print('USD EXCHANGE RATE')
display(usd_df)
```

```python
kospi_df_20250818 = kospi_df.loc[kospi_df['localDate'] >= '20250818']
kosdaq_df_20250818 = kosdaq_df.loc[kosdaq_df['localDate'] >= '20250818']
usd_df_20250818 = usd_df.loc[usd_df['localDate'] >= '20250818']
usd_df_20250818
```

```python
#    
kospi_date_set = set(kospi_df_20250818['localDate'])
kosdaq_date_set = set(kosdaq_df_20250818['localDate'])
usd_date_set = set(usd_df_20250818['localDate'])

common_dates = kospi_date_set.intersection(kosdaq_date_set,usd_date_set)

kospi_filtered = kospi_df_20250818.loc[kospi_df_20250818['localDate'].isin(common_dates)].sort_values('localDate').reset_index(drop=True)
kosdaq_filtered = kosdaq_df_20250818.loc[kosdaq_df_20250818['localDate'].isin(common_dates)].sort_values('localDate').reset_index(drop=True)
usd_filtered = usd_df_20250818.loc[usd_df_20250818['localDate'].isin(common_dates)].sort_values('localDate').reset_index(drop=True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 10))

plt.plot(kospi_filtered["localDate"], kospi_filtered["closePrice"], "r*:", label="kospi")
plt.plot(kosdaq_filtered["localDate"], kosdaq_filtered["closePrice"], "bo:", label="kosdaq")
plt.plot(usd_filtered["localDate"], usd_filtered["closePrice"], "g^--", label="usd")
plt.xticks(rotation=45, fontsize=7)
plt.title('KOSPI, KOSDAQ, USD Close Prices (Common Dates)')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import minmax_scale

plt.figure(figsize=(20, 8))

plt.plot(kospi_filtered["localDate"], minmax_scale(kospi_filtered["closePrice"]), "r*:", label="kospi")
plt.plot(kosdaq_filtered["localDate"],minmax_scale(kosdaq_filtered["closePrice"]), "bo:", label="kosdaq")
plt.plot(usd_filtered["localDate"],   minmax_scale(usd_filtered["closePrice"]), "g^--", label="usd")
plt.xticks(rotation=45, fontsize=7)
plt.title('MinMax Scaled KOSPI, KOSDAQ, USD Close Prices (Common Dates)')
plt.legend()
plt.show()
```

```python
headers = {
    "User-Agent": "Thunder Client (https://www.thunderclient.com)"
}
MUSINSA_API = 'https://api.musinsa.com/api2/hm/web/v5/pans/ranking/sections/199?storeCode=musinsa&gf=M&ageBand=AGE_BAND_20&period=MONTHLY&eventPeriod=BASIC_REALTIME&categoryCode=002000'
res = requests.get(MUSINSA_API,headers=headers)
datas = res.json()['data']['modules']
valid = datas[1:]

items = [ j['info'] for i in valid for j in i['items'] ]

# for i in valid:
#   for j in i["items"]:
#     items.append(j['info'])

items
```

```python
m_20 = pd.DataFrame(items)
m_20
```

```python
m_20 = m_20.drop(['strikethrough','onClickBrandName','optionPriceInformation','title','subTitle'],axis=1)
m_20
```

```python
m_20.info()
```

```python
m_20_brand = m_20.groupby('brandName')[['productName']].count()
m_20_brand = m_20_brand.rename(columns={'productName':"count"})
m_20_brand = m_20_brand.sort_values('count',ascending=False)
print('  >  >    :  | 20-24 | 1')
print(display(m_20_brand))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,9))
plt.bar(
    m_20_brand.index,
  m_20_brand['count'],

)
plt.title('1 20-24    ')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
