```python
import json

def url_lists(raw_text):

  # 2. 앞부분의 보안 기호 )]}', 와 줄바꿈 제거
  clean_json_text = raw_text.split('\n', 1)[1] if raw_text.startswith(')]}') else raw_text

  # 3. JSON 파싱
  data = json.loads(clean_json_text)

  # 4. searchList에서 postUrl만 추출
  post_urls = [post['postUrl'] for post in data['result']['searchList']]

  # 결과 출력
  for url in post_urls:
      print(url)

  return post_urls
```
