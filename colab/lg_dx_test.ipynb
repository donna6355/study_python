```python
#Q3 EMP 테이블을 활용하여 재직중인 대상 중에 성별이 남성인 팀장급만 아래와 같이
# 출력하세요. 단, 직원이름을 기준으로 오름차순 정렬하여 출력하세요.

SELECT NAME, RETIRE_YN, GENDER, JOB_TITLE
FROM EMP
WHERE RETIRE_YN = 'N'
AND GENDER = 'M'
AND JOB_TITLE = '팀장'
ORDER BY NAME
```

```python
# Q4 PRODUCT_REVIEW 테이블을 활용하여 각 상품별로 리뷰가 5 점인 대상이 몇 건
# 있는지 집계하세요. 단, 상품 ID 기준으로 오름차순 정렬하세요.

SELECT PRODUCT_ID, COUNT(*) AS 리뷰5점개수
FROM PRODUCT_REVIEW
WHERE RATING = 5
GROUP BY PRODUCT_ID
ORDER BY PRODUCT_ID ASC
```

```python
# PRODUCT_REVIEW 테이블을 활용해서 연도별로 2 점 이하의 리뷰가 몇 건
# 있었는지 아래와 같이 출력하세요. 단, 연도를 기준으로 오름차순 정렬하여 나오도록
# 해주세요.


SELECT TO_CHAR(CREATED_AT,'YYYY') AS 리뷰작성연도, COUNT(*) AS 리뷰개수2점이하만
FROM PRODUCT_REVIEW
WHERE RATING <=2
GROUP BY TO_CHAR(CREATED_AT,'YYYY')
ORDER BY TO_CHAR(CREATED_AT,'YYYY')
```

```python
# Q6 DEPT, EMP 테이블을 활용, 서브쿼리 기법을 이용해 부서별 인원수를 부서명과 함께
# DEPT_NAME 을 기준으로 오름차순 정렬하세요. (부서명은 중복되지 않는다고
# 가정하며 전체 직원을 대상으로 합니다.)
# * 서브쿼리를 사용하지 않아도 조인을 활용해 아래 와 동일한 결과가 출력된다면
# 정답으로 인정합니다.


SELECT D.DEPT_NAME, COUNT(E.EMP_ID) AS 부서별인원수
FROM DEPT D JOIN EMP E
ON D.DEPT_NO = E.DEPT_NO
GROUP BY D.DEPT_NAME
ORDER BY D.DEPT_NAME
```

```python
# Q7 아래 요구사항에 맞게 테이블을 생성하는 쿼리를 작성하세요.
# [요구사항]
# 회원의 인적사항을 저장할 테이블을 만들려고 합니다. 테이블명은 MEMBER_EXAM
# 입니다. 입력 받을 정보는 회원ID, 이름, 주민등록번호, 가입일시입니다. 또한 여기서는
# 위 한글명 그대로 활용해 테이블을 만들면 됩니다.
# 회원ID는 문자형으로 20BYTE까지 입력 받도록 하며, PK(식별자)로 설정합니다.
# 이름은 문자형으로 50BYTE까지 입력 받도록 하며, 필수로 입력 받아야 합니다.
# (null허용x) 주민등록번호는 문자형으로 255BYTE까지 입력 받도록 하며, 필수
# 입력은 아닙니다. 가입일시는 날짜형으로 입력 받고 따로 입력하지 않으면
# 현재시점(SYSDATE)이 입력되도록 기본값을 설정해줍니다.


CREATE TABLE MEMBER_EXAM (
  회원ID      VARCHAR2(20) PRIMARY KEY,
  이름        VARCHAR2(50) NOT NULL,
  주민등록번호  VARCHAR2(255),
  가입일시     DATE DEFAULT SYSDATE
)

```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
def get_chrome_driver():
    # 1. 크롬 옵션 세팅
    chrome_options = webdriver.ChromeOptions()

    # 2. driver 생성
    driver = webdriver.Chrome(
        service=Service(ChromeDriverManager().install()),
        options=chrome_options
    )

    return driver
```

```python
driver = get_chrome_driver()
driver.get("https://www.lge.co.kr/category/home-audio")

# products = driver.find_elements(By.CSS_SELECTOR, "#container > section.PlpPcContainer_container__L8y3U > section.PlpPcContainer_list_cont__tBF9C > section.CommonPcListUnitProduct_list_unit_product__AeuOT > ul > li")

page = driver.page_source
page
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page, "html.parser")

products = soup.select('#container > section.PlpPcContainer_container__L8y3U > section.PlpPcContainer_list_cont__tBF9C > section.CommonPcListUnitProduct_list_unit_product__AeuOT > ul > li')

def set_data(p):
  rating = p.select_one('.product-card-title_score').text
  name = p.select_one('.product-card-title_headline').text
  model = p.select_one('.product-card-title_sku').text
  return {
      "상품명": name,
      "평점" : rating,
      "모델명":model
  }

product_data = [set_data(p) for p in products]

product_df = pd.DataFrame(product_data)
product_df

```
