```python
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

# -----------------------------
# 1. 검색 URL 생성
# -----------------------------
keyword = '%22식기세척기%22%20%22테트리스%22%20%2B내돈내산'
startdate = '20200101'
enddate = '20260201'

blog_search_url = (
    f'https://search.naver.com/search.naver'
    f'?ssc=tab.blog.all'
    f'&query={keyword}'
    f'&sm=tab_opt'
    f'&nso=so%3Ar%2Cp%3Afrom{startdate}to{enddate}'
)

# -----------------------------
# 2. 드라이버 실행
# -----------------------------
driver = wb.Chrome()
driver.get(blog_search_url)
time.sleep(2)

# -----------------------------
# 3. 스크롤 내리기
# -----------------------------
body = driver.find_element(By.TAG_NAME, 'body')

for _ in range(5):
    body.send_keys(Keys.END)
    time.sleep(1)

# -----------------------------
# 4. iframe 자동 탐색 후 진입
# -----------------------------
iframes = driver.find_elements(By.TAG_NAME, "iframe")

for iframe in iframes:
    driver.switch_to.frame(iframe)
    links = driver.find_elements(By.CSS_SELECTOR, "a")
    if len(links) > 0:
        break
    driver.switch_to.default_content()

# -----------------------------
# 5. 블로그 글 URL 수집
# -----------------------------
href_list = []

links = driver.find_elements(By.CSS_SELECTOR, "a")

for el in links:
    href = el.get_attribute("href")
    if href and "blog.naver.com" in href:
        href_list.append(href)

# 중복 제거
href_list = list(set(href_list))

# -----------------------------
# 6. 결과 확인
# -----------------------------
print(f"수집된 블로그 글 수: {len(href_list)}")
href_list[:10]
```

```python
from selenium.webdriver.common.by import By
import time

negative_keywords = ["불편", "단점"]

negative_blog_urls = []   # 단점/불편 언급한 글
checked_blog_urls = []    # 정상 처리된 글

for idx, url in enumerate(href_list):
    try:
        print(f"[{idx+1}/{len(href_list)}] 접속 중:", url)
        driver.get(url)
        time.sleep(2)

        # -----------------------------
        # 1. iframe 진입
        # -----------------------------
        driver.switch_to.default_content()
        iframes = driver.find_elements(By.TAG_NAME, "iframe")

        for iframe in iframes:
            driver.switch_to.frame(iframe)
            if driver.find_elements(By.TAG_NAME, "body"):
                break
            driver.switch_to.default_content()

        # -----------------------------
        # 2. 본문 텍스트 수집
        # -----------------------------
        content_elements = driver.find_elements(
            By.CSS_SELECTOR,
            "div.se-main-container, div#postViewArea"
        )

        if not content_elements:
            print("  ❌ 본문 없음")
            continue

        content_text = content_elements[0].text

        # -----------------------------
        # 3. 키워드 포함 여부 확인
        # -----------------------------
        if any(keyword in content_text for keyword in negative_keywords):
            negative_blog_urls.append(url)
            print("  ⚠️ 단점/불편 키워드 발견")

        checked_blog_urls.append(url)

    except Exception as e:
        print("  ❌ 오류 발생:", e)
        continue

print("\n✅ 분석 완료")
print("단점/불편 언급 글 수:", len(negative_blog_urls))
```

```python
import re
from selenium.webdriver.common.by import By
import time

negative_keywords = ["불편", "단점", "어렵", "귀찮", "힘들", "아쉽"]

negative_sentences = []

for url in negative_blog_urls:
    driver.get(url)
    time.sleep(2)

    driver.switch_to.default_content()
    iframes = driver.find_elements(By.TAG_NAME, "iframe")

    for iframe in iframes:
        driver.switch_to.frame(iframe)
        if driver.find_elements(By.TAG_NAME, "body"):
            break
        driver.switch_to.default_content()

    content = driver.find_elements(
        By.CSS_SELECTOR,
        "div.se-main-container, div#postViewArea"
    )

    if not content:
        continue

    text = content[0].text

    sentences = re.split(r'[.!?\n]', text)

    for sent in sentences:
        if any(k in sent for k in negative_keywords):
            negative_sentences.append(sent.strip())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
```

```python
def extract_ngram_candidates(sentences, min_df=3):
    vectorizer = CountVectorizer(
        ngram_range=(2, 3),
        token_pattern=r"[가-힣]{2,}",
        min_df=min_df
    )
    
    X = vectorizer.fit_transform(sentences)
    ngrams = vectorizer.get_feature_names_out()
    
    return ngrams
```

```python
ngram_candidates = extract_ngram_candidates(negative_sentences, min_df=3)
len(ngram_candidates)
ngram_candidates
```

```python
from collections import defaultdict

def expand_complaint_patterns(base_patterns, ngram_candidates):
    expanded = defaultdict(set)
    
    for category, keywords in base_patterns.items():
        # 기존 키워드 유지
        for k in keywords:
            expanded[category].add(k)
        
        # n-gram 중 유사 표현 자동 추가
        for ng in ngram_candidates:
            for k in keywords:
                if k.replace(" ", "") in ng.replace(" ", ""):
                    expanded[category].add(ng)
    
    # set → list 변환
    return {k: sorted(list(v)) for k, v in expanded.items()}
```

```python
expanded_patterns = expand_complaint_patterns(
    complaint_patterns,
    ngram_candidates
)
```

```python
def get_new_keywords(original, expanded):
    new_only = {}
    for cat in expanded:
        new_only[cat] = list(
            set(expanded[cat]) - set(original.get(cat, []))
        )
    return new_only

new_keywords = get_new_keywords(complaint_patterns, expanded_patterns)

for cat, kws in new_keywords.items():
    print(f"\n[{cat} - 자동 추천]")
    for kw in kws:
        print(" +", kw)
```

```python
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import re
import pandas as pd

# 검색 url 생성
keyword = '%22식기세척기%22%20%2B내돈내산'
startdate = '20250101'
enddate = '20260201'

search_url = (
    f'https://search.naver.com/search.naver'
    f'?ssc=tab.blog.all'
    f'&query={keyword}'
    f'&sm=tab_opt'
    f'&nso=so%3Ar%2Cp%3Afrom{startdate}to{enddate}'
)

driver = wb.Chrome()
driver.get(search_url)
time.sleep(2)

# 검색 결과 스크롤
body = driver.find_element(By.TAG_NAME, 'body')

for _ in range(5):
    body.send_keys(Keys.END)
    time.sleep(1)

# 공통 함수 : 블로그 iframe 진입
def switch_to_blog_iframe(driver):
    driver.switch_to.default_content()
    for iframe in driver.find_elements(By.TAG_NAME, "iframe"):
        driver.switch_to.frame(iframe)
        if driver.find_elements(By.TAG_NAME, "body"):
            return
        driver.switch_to.default_content()

# 블로그 글 URL 수집
switch_to_blog_iframe(driver)

href_list = []
links = driver.find_elements(By.CSS_SELECTOR, "a")

for link in links:
    href = link.get_attribute("href")
    if href and "blog.naver.com" in href:
        href_list.append(href)

href_list = list(set(href_list))
print(f"수집된 블로그 글 수: {len(href_list)}")


# 불편 키워드 및 카테고리 정의
negative_keywords = ["불편", "단점", "어렵", "귀찮", "힘들", "아쉽"]

complaint_categories = {
    "테트리스": ["테트리스", "쌓", "배치", "넣기 어렵"],
    "애벌세척": ["애벌", "미리 씻", "헹궈"],
    "소음": ["소음", "시끄럽", "소리"],
    "시간": ["시간", "오래", "느리"],
    "용량": ["용량", "작", "적다"]
}

# 블로그 1회 접속 -> 불편 문장 추출 + 분류
rows = []

for idx, url in enumerate(href_list):
    try:
        print(f"[{idx+1}/{len(href_list)}] 분석 중:", url)
        driver.get(url)
        time.sleep(2)

        switch_to_blog_iframe(driver)

        content = driver.find_elements(
            By.CSS_SELECTOR,
            "div.se-main-container, div#postViewArea"
        )

        if not content:
            continue

        text = content[0].text
        sentences = re.split(r'[.!?\n]', text)

        for sent in sentences:
            sent = sent.strip()
            if not sent:
                continue

            # 1차: 불편 키워드 필터
            if not any(k in sent for k in negative_keywords):
                continue

            # 2차: 불만 유형 분류
            for category, keywords in complaint_categories.items():
                if any(k in sent for k in keywords):
                    rows.append({
                        "url": url,
                        "category": category,
                        "sentence": sent
                    })

    except Exception as e:
        print("❌ 오류 발생:", e)

df_result = pd.DataFrame(rows)
df_result.to_csv(
    "식기세척기_불편요소_카테고리분석.csv",
    index=False,
    encoding="utf-8-sig"
)

print("✅ 분석 완료")
print(df_result["category"].value_counts())


# 결과 데이터프레임 생성 및 저장
df_result = pd.DataFrame(rows)
df_result.to_csv(
    "식기세척기_불편요소_카테고리분석.csv",
    index=False,
    encoding="utf-8-sig"
)

print("✅ 분석 완료")
print(df_result["category"].value_counts())
```
