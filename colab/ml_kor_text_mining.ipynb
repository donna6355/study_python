```python
import nltk
nltk.download("punkt")
nltk.download("punkt_tab")
nltk.download("averaged_perceptron_tagger_eng")
```

```python
#tokenizing 의미를 나타내는 가장 작은 단위로 쪼개는 것
#sample

text = 'Good morning Isaac the cutest cat in the world'
```

```python
split_text= nltk.word_tokenize(text)
split_text
```

```python
#품사정보 taggin
#중요도 : 명사 > 형용사/동사 > 부사
tagged_text = nltk.pos_tag(split_text)
tagged_text
```

```python
# NNP, NN, NNS, NNPS
sample = ['NNP',"NN", "NNS", "NNPS"]
nouns = []
for text, form in tagged_text:
  if form in sample:
    nouns.append(text)
nouns
```

```python
import pandas as pd
df = pd.read_csv('./data/kor_crawling.csv',encoding='utf-8-sig')
df
```

```python
df['Review'] = df['Review'].apply(lambda x : x.lower())
df
```

```python
df['length'] = df['Review'].apply(lambda x : len(x))
df = df.drop(df[df['length'] <12].index).reset_index(drop=True)
del df['length']
df
```

```python
from kiwipiepy import Kiwi
from kiwipiepy.utils import Stopwords
stopwords = Stopwords()
kiwi = Kiwi()



df['token'] = df['Review'].apply(lambda x:[i[0] for i in kiwi.tokenize(x,stopwords=stopwords)])
df
```

```python
okt = Okt()

df['token'] = df['Review'].apply(lambda x: okt.morphs(x,stem=True,norm=True))
df
```

```python
sw = [i[0] for i in stopwords.stopwords]
sw
```

```python
cleaned =[]
for i in df['token']:
  imsi = [ w for w in i if w not in sw]
  # imsi = []
  # for w in i:
  #   if w not in sw:
  #     imsi.append(w)
  cleaned.append(imsi)

df['token'] = cleaned
df
```

```python
#LDA
from gensim import corpora, models
import gensim
```

```python
lda = list(df['token'])
dictionary = corpora.Dictionary(lda)
corpus = [dictionary.doc2bow(text) for text in lda]
corpus[0:2] # (BOW num, frequency)

topic_num = 3
model = gensim.models.ldamodel.LdaModel(corpus, num_topics=topic_num, id2word=dictionary)

```

```python
model.print_topics(num_words=5)
```

```python
model.get_document_topics(corpus)[0] #0번째 리뷰가 어떤 토픽에 할당될 확률이 높은지
```

```python
import numpy as np


align =[]
for i in model.get_document_topics(corpus):
  label = []
  value = []
  for l, p in i:
    label.append(l)
    value.append(p)
  align.append(label[np.argmax(value)])
df['topic'] = align
df
```
