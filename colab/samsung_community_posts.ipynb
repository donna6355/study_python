```python
#삼성 커뮤니티 활성화를 알아보기 위해서 
#하루 (현재 기준 어제 하루)에 올라오는 포스트 수, 좋아요, 조회 수 등을 수집하는 코드입니다. 
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from datetime import datetime, timedelta
import pandas as pd
import time 
```

```python

def get_chrome_driver():
  #chrome option setting
  chrome_options = webdriver.ChromeOptions()

  #create driver
  driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=chrome_options
  )

  return driver
```

```python
driver = get_chrome_driver()
```

```python
SAMSUNG ='https://r1.community.samsung.com/t5/korea/ct-p/kr?page=1&tab=recent_topics'
driver.get(SAMSUNG)
time.sleep(3)
```

```python
# 더 불러오기 클릭
def click_more():
    driver.find_element(By.CSS_SELECTOR, "#lia-body > div.lia-page > center > div > div > div > div > div.lia-quilt.lia-quilt-category-page-country-landing-page.lia-quilt-layout-samsung-basic-category.lia-top-quilt > div.lia-quilt-row.lia-quilt-row-full-width-content > div > div > div.samsung-community-activity > section > div.lia-view-all > a").click()

click_more()
```

```python
# 마지막 포스트 날짜 확인하고 오늘이면 계속 더불러오기를 클릭하여 오늘 작성된 

def check_date():
    post_date = driver.find_element(By.CSS_SELECTOR,'#lazy-loaded-content > article:last-child > header > div.author > time').text
    print('last post date',post_date)
    return post_date
check_date()
```

```python

# 오늘 / 어제 날짜
today = datetime.now().date()
yesterday = today - timedelta(days=1)

posts = []

while True:
    # 마지막 포스트 날짜 가져오기
    last_date_text =check_date()

    last_date = datetime.strptime(last_date_text, "%m-%d-%Y %I:%M %p").date()

    print("현재 마지막 포스트 날짜:", last_date)

    # 오늘 또는 어제면 더 불러오기
    if last_date >= yesterday:
        try:
            click_more()
            time.sleep(3)
        except:
            print("더 불러오기 버튼 없음")
            break
    else:
        print("어제 이전 날짜 발견 → 중단")
        break
```

```python
page = driver.page_source
page
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page, "html.parser")

posts = soup.select('.samsung-message-tile')
posts
```

```python
posts[0]
```

```python
def forge_data(post):
    created_at = post.select_one('time').text
    content = post.select_one('.content p').text
    view = post.select_one('.samsung-tile-views b').text
    reply = post.select_one('.samsung-tile-replies b').text
    like = post.select_one('.samsung-tile-kudos b').text
    return {
        "created_at":created_at,
        'content':content,
        'view':view,
        'reply':reply,
        'like':like
    }
posts_dict = [ forge_data(post) for post in posts]
posts_dict
```

```python
df = pd.DataFrame(posts_dict)
df
```

```python
df["created_at"] = pd.to_datetime(df["created_at"], format="%m-%d-%Y %I:%M %p")
df.info()
```

```python
#어제 하루 날짜 데이터만 거르기
yesterday = (datetime.now() - timedelta(days=1)).date()
df_yesterday = df[df["created_at"].dt.date == yesterday]
df_yesterday
```

```python
df_yesterday['view'] = df_yesterday['view'].astype(int)
df_yesterday['reply'] = df_yesterday['reply'].astype(int)
df_yesterday['like'] = df_yesterday['like'].astype(int)
df_yesterday.info()
```

```python
#엑셀 파일로 저장하기
df_yesterday.to_excel("samsung_recent_posts.xlsx", index=False)
```
