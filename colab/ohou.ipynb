```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from datetime import datetime, timedelta
import pandas as pd
import time 

# -------------------------
#  ë“œë¼ì´ë²„ ì„¤ì •
# -------------------------
def get_chrome_driver():
  #chrome option setting
  chrome_options = webdriver.ChromeOptions()

  #create driver
  driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=chrome_options
  )

  return driver

driver = get_chrome_driver()

OHOU ='https://contents.ohou.se/community/product'
driver.get(OHOU)
time.sleep(3)

driver.execute_script("window.scrollBy(0, 1000);")
# -------------------------
#  1ë…„ ì „ ë‚ ì§œ ê³„ì‚°
# -------------------------
one_year_ago = datetime.now() - timedelta(days=365)



collected = []
seen_keys = set()
stop_flag = False
```

```python
post_sel = '#__next > div.css-zf0iqh.etzpybq0 > div.css-15y92lb.etzpybq1 > div > div:nth-child(3) > div > div > div > div'
category_sel = 'div > span > a > div > div.css-fcukdp.eh1ttt1 > div > div.css-1ytovu8 > div.css-j7qwjs > div.css-kuvu0c > div > span'
title_sel = 'div > span > a > div > div.css-fcukdp.eh1ttt1 > div > div.css-1ytovu8 > div.css-j7qwjs > div.e1868rkl0.css-h7lda > span'
view_sel = 'div > span > a > div > div.css-fcukdp.eh1ttt1 > div > div.css-1oc4nvw > div:nth-child(1) > span:nth-child(5)'
created_sel = 'div > span > a > div > div.css-fcukdp.eh1ttt1 > div > div.css-1oc4nvw > div:nth-child(1) > span:nth-child(3)'
```

```python
posts = driver.find_elements(By.CSS_SELECTOR, post_sel)
test = posts[1].find_element(By.CSS_SELECTOR, category_sel).text.strip().replace('ì¡°íšŒ ','')
print(test)
test = posts[1].find_element(By.CSS_SELECTOR, title_sel).text.strip().replace('ì¡°íšŒ ','')
print(test)
test = posts[1].find_element(By.CSS_SELECTOR, view_sel).text.strip().replace('ì¡°íšŒ ','')
print(test)
test = posts[1].find_element(By.CSS_SELECTOR, created_sel).text.strip().replace('ì¡°íšŒ ','')
test
```

```python

# ------------------------
#  ê°€ìƒ ìŠ¤í¬ë¡¤ ìˆ˜ì§‘
# ------------------------


while not stop_flag:
    
    time.sleep(1)
    
    posts = driver.find_elements(By.CSS_SELECTOR, post_sel)
    
    for post in posts:
        try:
            post_idx = post.get_attribute("data-index")
            if(post_idx in seen_keys):
                print(f'ì¤‘ë³µ í¬ìŠ¤íŠ¸ ìŠ¤í‚µ! {post_idx}')
                break

            seen_keys.add(post_idx)
            # ë‚ ì§œ
            date_el = post.find_element(By.CSS_SELECTOR, created_sel).text.strip()
            
            # ì œëª©
            title = post.find_element(By.CSS_SELECTOR, title_sel).text.strip()

            # ì¹´í…Œê³ ë¦¬
            category = post.find_element(By.CSS_SELECTOR, category_sel).text.strip()

            # ì¡°íšŒìˆ˜ / ëŒ“ê¸€ìˆ˜
            view = post.find_element(By.CSS_SELECTOR, view_sel).text.strip().replace('ì¡°íšŒ ','')
            print(f'{post_idx}ìˆ˜ì§‘ {title}')
            
            collected.append({
                "ì‘ì„±ë‚ ì§œ": date_el,
                "ì¹´í…Œê³ ë¦¬": category,
                "ì œëª©": title,
                "ì¡°íšŒìˆ˜": view,
            })
        except:
            print('failed to collect data')
            continue
    if (len(collected) >10000):
        break
    # ìŠ¤í¬ë¡¤ ì¡°ê¸ˆì”©
    driver.execute_script("window.scrollBy(0, 1000);")
    time.sleep(2)


print(f"ìˆ˜ì§‘ ì™„ë£Œ: {len(collected)}ê°œ")
driver.quit()
```
