### Combine Multiple CSV Files

This section will demonstrate how to combine multiple CSV files with identical columns into a single DataFrame and then save it as a new CSV file.

```python
import pandas as pd
import glob
import os
```

First, define the directory where your CSV files are located and the pattern to match them (e.g., `*.csv`).

```python
# Define the path to your CSV files
csv_directory = './all' # <---- IMPORTANT: Change this to your directory

# Get a list of all CSV files in the directory
all_files = glob.glob(os.path.join(csv_directory, '*.csv'))

# Display the list of files found
print(f"Found {len(all_files)} CSV files:")
for f in all_files:
    print(f)
```

Next, we will iterate through each file, read it into a pandas DataFrame, and append it to a list. Finally, we'll concatenate all DataFrames in the list into one master DataFrame.

```python
li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

# Concatenate all dataframes into one
combined_df = pd.concat(li, axis=0, ignore_index=True)

# Display the first few rows of the combined DataFrame
print("Combined DataFrame head:")
display(combined_df.head())
```

Finally, save the combined DataFrame to a new CSV file.

```python
combined_df.info()
```

```python
# Define the output path for the combined CSV file
output_csv_path = 'combined_data.csv' # <---- You can change the output filename here

# Save the combined DataFrame to a new CSV file
combined_df.to_csv(output_csv_path, index=False)

print(f"Successfully combined all CSV files into '{output_csv_path}'")
```
