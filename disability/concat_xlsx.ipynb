```python
import pandas as pd
import os

# Define the directory where your .xlsx files are located
data_directory = './data/'

# Get a list of all .xlsx files in the directory
xlsx_files = [f for f in os.listdir(data_directory) if f.endswith('.xlsx')]

# Create an empty list to store dataframes
df_list = []

# Loop through each .xlsx file, read it into a DataFrame, and append to the list
for file in xlsx_files:
    file_path = os.path.join(data_directory, file)
    print(f"Reading {file_path}...")
    try:
        df = pd.read_excel(file_path)
        df_list.append(df)
    except Exception as e:
        print(f"Error reading {file_path}: {e}")

# Concatenate all DataFrames in the list into a single DataFrame
if df_list:
    combined_df = pd.concat(df_list, ignore_index=True)

    # Define the output CSV file path
    output_csv_path = os.path.join(data_directory, 'combined_reviews.csv')

    # Save the combined DataFrame to a CSV file
    combined_df.to_csv(output_csv_path, index=False, encoding='utf-8-sig')

    print(f"Successfully combined {len(xlsx_files)} Excel files into {output_csv_path}")
    print("First 5 rows of the combined CSV:")
    display(combined_df.head())
else:
    print("No .xlsx files found in the directory or an error occurred while reading them.")
```

```python
# Convert '리뷰날짜' to datetime objects
combined_df['리뷰날짜'] = pd.to_datetime(combined_df['리뷰날짜'])

# Sort the DataFrame by '리뷰날짜'
combined_df_sorted = combined_df.sort_values(by='리뷰날짜').reset_index(drop=True)

print("DataFrame sorted by '리뷰날짜':")
display(combined_df_sorted.head())
```

```python
date_filter = '2020-01-01'
filtered_reviews = combined_df_sorted[combined_df_sorted['리뷰날짜'] > date_filter]

print(f"Number of reviews after {date_filter}: {len(filtered_reviews)}")
display(filtered_reviews.head())
```
