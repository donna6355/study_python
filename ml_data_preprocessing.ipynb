```python
#1. 결측치를 지운다
df = df.dropna()
df
```

```python
np.mean(df['pH'])
df.columns
```

```python
#2. 결측치를 적절한 숫자로 채운다 - 평균
for i in df.columns:
    df[i].loc[6298] = np.mean(df[i])
df
```

```python
#스케일링
```

```python
mean = np.mean(df['citric acid'])
std = np.std (df['citric acid'])
scaled_acid = []

for i in df['citric acid']:
    scaled_acid.append((i-mean)/std)
scaled_acid
```

```python
maxx = np.max(df['citric acid'])
minn = np.min(df['citric acid'])
scaled_acid = []

for i in df['citric acid']:
    scaled_acid.append((i-minn)/(maxx-minn))
```
