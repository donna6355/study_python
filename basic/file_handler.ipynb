```python
from PIL import Image
import os
```

```python
def process_directory(input_dir, output_dir):


    if not os.path.exists(output_dir):
        os.makedirs(output_dir)


    supported_formats = ('.png')


    for filename in os.listdir(input_dir):
        if filename.lower().endswith(supported_formats):
            input_path = os.path.join(input_dir, filename)

            output_filename = os.path.splitext(filename)[0] + '.webp'
            output_path = os.path.join(output_dir, output_filename)
            convert_to_webp(input_path, output_path)
```

```python
def convert_to_webp(input_path, output_path):

    try:
        image = Image.open(input_path)


        if image.mode in ('RGBA', 'LA'):
            background = Image.new('RGB', image.size, (255, 255, 255))
            background.paste(image, mask=image.split()[-1])
            image = background

        image.save(output_path, 'WEBP', quality=80)

        print(f"변환 성공: {input_path} -> {output_path}")

    except Exception as e:
        print(f"변환 실패: {input_path}, 에러: {str(e)}")
```

```python
org = './ym'
cvt = './converted'
process_directory(org, cvt)
```

```python
def rename_files_with_increment(directory, prefix='ym_', start_num=1, extension='.webp'):
    """
    Renames files in a given directory with an incrementing number.

    Args:
        directory (str): The path to the directory containing the files.
        prefix (str): The prefix for the new file names (e.g., 'image_').
        start_num (int): The starting number for the incrementing sequence.
        extension (str): The extension of the files to rename (e.g., '.webp').
    """
    if not os.path.exists(directory):
        print(f"Error: Directory '{directory}' does not exist.")
        return

    # List files and filter by extension
    files_to_rename = [f for f in os.listdir(directory) if f.lower().endswith(extension)]
    files_to_rename.sort() # Sort to ensure consistent renaming order

    print(f"Found {len(files_to_rename)} '{extension}' files to rename in '{directory}'.")

    for i, filename in enumerate(files_to_rename):
        old_path = os.path.join(directory, filename)
        new_filename = f"{prefix}{start_num + i}{extension}"
        new_path = os.path.join(directory, new_filename)

        try:
            os.rename(old_path, new_path)
            print(f"Renamed '{filename}' to '{new_filename}'")
        except Exception as e:
            print(f"Error renaming '{filename}': {e}")

    print("File renaming complete.")
```

Here's an example of how to use the `rename_files_with_increment` function. You can change `target_directory`, `new_prefix`, `starting_number`, and `file_extension` as needed.

```python
# Example usage:
target_directory = './converted' # Or any other directory you want to rename files in
new_prefix = 'ym_'
starting_number = 1
file_extension = '.webp' # Make sure this matches the files you want to rename

rename_files_with_increment(target_directory, new_prefix, starting_number, file_extension)
```
