# 1. HTML
- HyperText Markup Language
  - HyperText : 페이지의 이동이 가능한
  - Markup : 구간이 있는 (표식)
  - Language : 언어
- Markup
  - 나는 어제 사과를 먹었다.
    - 사람은 쉽게 이해할 수 있으나, 컴퓨터는 이해하기가 힘들어요.
  - Markup 예시
  -  ```html
        <누가>나</누가>
        <언제>어제</언제>
        <무엇을>사과</무엇을>
        <어떻게>먹었다</어떻게>
    ```

- HTML은 웹에 보이는 화면인 웹 페이지에 필요한 구성요소를 구성하는 Markup 언어
- 웹 페이지를 구성하는 언어
  - HTML : 마크업을 이용해 화면의 레이아웃 구조와 텍스트, 엘리먼트를 표현한다.
  - CSS(Cascade Style Sheet) : 엘리먼트를 꾸며주는 언어
    - **html 엘리먼트를 선택**을 한 다음 꾸며줘야 한다.
      - `selector`를 이용한 엘리먼트 선택
  - JavaScript : 사용자와 웹 페이지의 인터렉션을 담당

## HTML의 구성 요소
- Document
  - 눈에 보이는 페이지 전체
- Element
  - Document를 구성하는 **요소**
  - 레이아웃
    - 영역을 나눠주는 역할. 대표적으로 `<div>, <p>, <span> 등등`
  - 엘리먼트
    - 특정한 기능이나 표현이 있는 구성요소
    - 눈으로 볼 수 있는 요소
    - 레이아웃이 아니면 모두 엘리먼트
  - 엘리먼트는 **계층적 구조**로 이루어져 있다.
    - 엘리먼트 내에 다른 엘리먼트가 배치될 수 있다.
    - **DOM(Document Object Model)** ★★★★★
- Tag
  - 목적이 다른 태그들이 각각 존재한다.
  - 태그의 이름에 따라서 태그의 모양과 기능(목적)이 달라진다.
  - 태그들을 이용해 엘리먼트 구조를 만들어 낸다.
- Attribute
  - 엘리먼트의 속성을 표현 `attribute="attribute value"`
  - 시작 태그 안에 포함되는 속성값

````python
%%html
<button>HTML1</button>
````

속성 부여

````python
%%html
<button style="color:red;">HTML1</button>
````

## 속성값 종류
- `id, class` 속성
  - **엘리먼트를 선택**하기 위한 속성
  - `id`속성 : 전페 웹 페이지에서 유일한 값
  - `class`속성
    - 웹 페이지에서 여러 가지 값을 사용할 수 있다. **중복 가능**
    - 하나의 엘리먼트 내에 여러 개의 `class`를 설정 가능
- `기타 등등`
  - 엘리먼트의 기능 등을 담당
  - 엘리먼트의 크기, 배경 화면 등등 스타일 적인 요소를 담당
  - `a` 태그 같은 경우는 이동할 페이지를 입력


## `hn` 태그
- 문서 내 제목을 나타내기 위한 태그
- <h`n`> 제목 </h`n`>
  - `n`은 1 ~ 6 숫자
  - `n`이 커질수록 글씨가 작아진다.

````python
%%html
<h1>안녕하세요!</h1>
<h2>안녕하세요!</h2>
<h3>안녕하세요!</h3>
<h4>안녕하세요!</h4>
<h5>안녕하세요!</h5>
<h6>안녕하세요!</h6>
````

# `p` 태그
- Paragraph
- 문단을 표현하기 위한 태그
- `<p>문단에 들어갈 내용들~~~</p>`

````python
%%html
<p>하고싶은 아무말이나 적으면 됩니다. 여기는 문단이 들어가는 공간이에요~</p>
<p>여기는 두 번째 문단입니다. 오늘 내일 다음주까지 크롤링을 배울거에요~</p>
````

- 문자열을 일부만 감싸고 있는 태그
- 문자열의 특정한 부분을 감싸서 **스타일**을 적용해 줄 수 있다.

# `table` 태그
- 표를 그리기 위한 태그 (데이터 프레임)
    * `table` : 전체 테이블을 만들기 위한 태그
    * `caption` : 표의 제목을 만들기위한 태그(생략 가능)
    * `thead` : 테이블을 항목(컬럼)을 표시하기위한 태그 (생략 가능)
    * `tbody` : 테이블에 표시할 실제 데이터가 들어가는 태그 (생략 가능)
    * `th` : 컬럼명을 표기하기 위한 태그
    * `td` : 데이터를 표기하기 위한 태그
    * `tr` : 한 행을 만들기 위한 태그


````python
%%html
<table border="1">
  <caption>테이블 제목</caption>
  <thead>
    <tr>
      <th>코드</th>
      <th>회사이름</th>
    </tr>
  </thead>
  <tbody>
    <tr>
      <td>00001</td>
      <td>삼성전자</td>
    </tr>
    <tr>
      <td>00002</td>
      <td>하이닉스</td>
    </tr>
  </tbody>
</table>
````

# 리스트 태그
- `ul` (Unordered list)
  - 숫자가 없이 불릿으로만 나열되는 리스트
- `ol` (Ordered list)
  - 숫자를 이용해서 순서를 표현할 수 있는 리스트
- `li`
  - `ul` 또는 `ol` 등에 대한 리스트 항목을 만들 때 사용


````python
%%html

<ul>
  <li>data 1</li>
  <li>data 2</li>
  <li>data 3</li>
</ul>
````

````python
%%html

<ol>
  <li>data 1</li>
  <li>data 2</li>
  <li>data 3</li>
</ol>
````

# CSS Selector
- CSS : html 엘리먼트를 꾸미기 위한 시트
  - 엘리먼트를 선택자(selector)를 이용해 선택을 하고 꾸미게 됩니다.
- CSS 생성 방법
  - HTML 내에 `<style></style>`
  - 엘리먼트의 style attribute를 활용
    - `style="color:red; background=blue;"`
    - 선택자를 활용하지 않는 방법.

## Selector의 종류
- 태그의 이름을 이용해서 선택
  - `div, span, input, p, a, ul, li 등등..` 태그명을 이용해서 선택
  - 한 웹페이지에 여러 개의 태그가 있을 수 있기 때문에 **여러 개**의 엘리먼트가 선택
- `id` 속성을 이용해서 선택
  - 웹 페이지에 존재하는 유일한 속성을 이용해 엘리먼트를 선택
  - 1개의 엘리먼트만 선택
- `class` 속성을 이용해서 선택
  - 웹 페이지에 여러 개 존재하는 값을 이용해서 선택
  - **여러 개**의 엘리먼트를 선택
- `attribute` 속성을 이용해서 선택
  - 엘리먼트에 특정 효과를 부여하는 `attribute`를 이용해서 여러 개의 엘리먼트를 선택

## CSS 문법
```html
<style>
  selector {
    꾸미기 시트
  }
</style>
```

````python
%%html
<style>
  p{
    color:red
  }
  #sp {
    color:green
  }
  .cs{
    color:brown
  }
</style>


<div>Hello</div>
<p>Greetings</p>
<span class="cs">hihi</span>
<p id='sp'>Bye</p>
````

````python
#NAVER MARKET INDEX
import requests

NAVER = 'https://finance.naver.com/marketindex/'

res = requests.get(NAVER)
res
````

````python
page = res.content
page
````

````python
soup = BeautifulSoup(page, 'html.parser')
soup
````

````python
ex_list = soup.select('#exchangeList > li')
ex_list
````

````python
test_li = ex_list[0]
test_li
````

````python
nation = test_li.select_one('h3 > span').text
nation[:nation.index(" ")]
````

````python
rate = test_li.select_one('.value').text
rate
````

````python
diff = test_li.select_one('.change').text.strip()
diff
````

````python
status = test_li.select_one('.head_info > span:last-child').text
status
````

````python
import pandas as pd

def extract_info(info_list):
  return {
      "nation": info_list.select_one('h3 > span').text[:nation.index(" ")],
      "rate":info_list.select_one('.value').text,
      "diff":info_list.select_one('.change').text.strip(),
      "status":info_list.select_one('.head_info > span:last-child').text
  }

df_ex = pd.DataFrame([extract_info(li) for li in ex_list])
df_ex
````
