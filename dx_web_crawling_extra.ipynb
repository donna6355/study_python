```
from selenium import webdriver as wb # 브라우저 실행 및 정보를 보관하는 역할
from selenium.webdriver.common.by import By # 요소를 찾기(선택) 위한 도구를 제공하는 역할
from selenium.webdriver.common.keys import Keys # 컴퓨터 키보드의 값을 제공하는 역할

import time
```

```
# browser
URL = 'https://cafe.naver.com/remonterrace'
driver = wb.Chrome()
driver.maximize_window()
driver.get(URL)
```

```
#move to specific board
driver.find_element(By.ID,'menuLink23').click()
```

```
# view 50 stories in a page
driver.find_element(By.XPATH, '//*[@id="cafe_content"]/div[2]/div/div[3]/div/button').click()
driver.find_element(By.XPATH, '//*[@id="cafe_content"]/div[2]/div/div[3]/div[2]/ul/li[7]/button').click()
```

```
#search keyword and enter
INPUT = driver.find_element(By.CSS_SELECTOR, '#cafe_content > div.SearchBoxLayout.type_bottom > div.SearchBox > div.search_input_area > input')

INPUT.click()
INPUT.send_keys('음식물처리기'+Keys.ENTER)
#INPUT.send_keys(Keys.ENTER)
```

```
href_list = []

org_list = driver.find_elements(By.CSS_SELECTOR, 'a.article')
for i in org_list:
    href_list.append(i.get_attribute('href'))

href_list
```

```
def get_info(page):

    driver.get(page)
    driver.switch_to.frame('cafe_main')
    time.sleep(1)
    title = driver.find_element(By.CSS_SELECTOR, 'h3.title_text').text
    content = driver.find_element(By.CSS_SELECTOR, 'div.se-main-container').text
    date = driver.find_element(By.CSS_SELECTOR,'span.date').text
    return {'title':title,'content': content,'date': date}

get_info(href_list[0])
```

```
data_list = [get_info(i) for i in href_list[0:3]]
```

```
import pandas as pd
cafe_df = pd.DataFrame(data_list)
cafe_df
```

```
cafe_df.to_csv('naver_cafe_lemonterrace.csv', encoding="utf-8-sig")
```

```
driver.quit()
```
