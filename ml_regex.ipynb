```python
#wildcard for single character .
b = re.compile('배.파')
```

```python
b.match('배아ㅎ파')
```

```python
#wildcard for more than 0 times character
b = re.compile('배고*파')
print(b.match('배파'))
print(b.match('배고파'))
print(b.match('배고고고고파'))
```

```python
#wildcard for more than 1 times character
b = re.compile('배고+파')
print(b.match('배파'))
print(b.match('배고파'))
print(b.match('배고고고고파'))
```

```python
#wildcard for exist or non
b = re.compile('배고?파')
print(b.match('배파'))
print(b.match('배고파'))
print(b.match('배고고고고파'))
```

```python
#wildcard for n times character
b = re.compile('배고{2}파')
print(b.match('배파'))
print(b.match('배고파'))
print(b.match('배고고파'))
print(b.match('배고고고고파'))
```

```python
#wildcard for n~m times character
b = re.compile('배고{2,4}파')
print(b.match('배파'))
print(b.match('배고파'))
print(b.match('배고고파'))
print(b.match('배고고고고파'))
print(b.match('배고고고고고고고고파'))
```

```python
#블라인드 채용 보조 시스템을 만들고 있습니다.
#자기 소개서에 학교 이름을 기재하면 안됩니다.
#00대학교, 00대학, 00여대, 00공대
#와 같이, 대학 이름을 적은 경우를 매칭하는 정규 표현식을 작성
```

```python
uni = re.compile('.{2,10}여?공?대학?교?')
print(uni.match('이화여대'))
print(uni.match('군대'))
print(uni.match('서울시립대'))
```

```python
# wildcard to binding
c = re.compile('배(고파)+')

print(c.match('배파'))
print(c.match('배고파'))
print(c.match('배고고파'))
print(c.match('배고파고파고파'))
```

```python
# wildcard to each item
c = re.compile('배[고파]+')

print(c.match('배파'))
print(c.match('배고파'))
print(c.match('배고고파'))
print(c.match('배고파고파고파'))
print(c.match('배고파파파'))
```

```python
# ignore wildcard
c = re.compile('배\+고파')

print(c.match('배파'))
print(c.match('배고파'))
print(c.match('배고고파'))
print(c.match('배고파고파고파'))
print(c.match('배고파파파'))
```

```python
# except
c = re.compile('배[^고파]')

print(c.match('배파'))
print(c.match('배고파'))
print(c.match('배고고파'))
print(c.match('배고파고파고파'))
print(c.match('배고파파파'))
print(c.match('배불러'))
print(c.match('배불'))
print(c.match('배'))
```

```python
c = re.compile('[a-z]')
print(c.match('a'))
print(c.match('A'))
c = re.compile('[a-zA-Z]')
print(c.match('a'))
print(c.match('A'))
print(c.match('ZA'))
print(c.match('aza'))
c = re.compile('[a-zA-Z0-9]')
print(c.match('a'))
print(c.match('A'))
print(c.match('ZA'))
print(c.match('aza'))
print(c.match('123'))
c = re.compile('[a-zA-Z0-9ㄱ-ㅎ]')
print(c.match('ㅎ'))
c = re.compile('[a-zA-Z0-9ㄱ-ㅎ가-힣]')
print(c.match('희'))
```

```python
#블라인드 채용 보조 시스템을 만들고 있습니다.
#자기 소개서에 전화번호를 기재하면 안됩니다.
#010-1234-5678, 02-123-4567, 031)123-1234, 01012345678
num = re.compile(r'0[0-9]{1,2}\-?\)?[0-9]{3,4}\-?[0-9]{4}')
print(num.match('010-2340-5555'))
print(num.match('01)2340-5555'))
print(num.match('0123405555'))
print(num.match('01+23405555'))
```

```python
#여러분들은 어제의 뉴스기사로 오늘의 주가를 예측하는 모델을 만들 거에요
#그래서 기사를 매일 크롤링하는데, 기사의 끝에 기자의 이메일이 있습니다.
#newdlckdgus@naver.com, NEWDLCKDGUS@hanyang.ac.kr, goodgood1234@nrf.or.kr
email = re.compile(r'[a-zA-Z0-9]{2,20}\@[a-zA-Z]{2,20}\.[a-z\.]{2,100}')
print(email.match('newdl@naver.com'))
print(email.match('waiting00@ngs.co.kf'))
print(email.match('as@.gg.gg.gg'))
print(email.match('newdlckdgus@naver.com'))
print(email.match('NEWDLCKDGUS@hanyang.ac.kr'))
print(email.match('goodgood1234@nrf.or.kr'))
```

```python
# 여러분들은 유튜브, 넷플릭스 등을 지속적으로 크롤링하면서, 다음 주에 인기 있을 것 같은 신제품 아이디어를 만들어야 합니다.
# 다만 특히나 유튜브는 유해 링크 댓글이 많아요.
# http://www.youhae.com, www.notgood.co.kr, https://www.advertise.com

url = re.compile(r'(https?\:\/\/)?(www\.)?[a-zA-z0-9]+\.[a-zA-z0-9\.]+')
print(url.match('http://www.youhae.com'))
print(url.match('https://www.advertise.com'))
print(url.match('www.notgood.co.kr'))
print(url.match('www.co.kr'))
print(url.match('www@naver.co.kr'))
```

```python
# single digit
d = re.compile(r'\d')
d.match('1')
d.match('1111')
```

```python
# except digit
d = re.compile(r'\D')
print(d.match('@'))
print(d.match('a'))
print(d.match('1'))
```

```python
# empty space
# use to distiguish the start of new sentece
d = re.compile(r'\s')

print(d.match(' '))
print(d.match('a'))
print(d.match('1'))
```

```python
# except empty space
d = re.compile(r'\S')

print(d.match(' '))
print(d.match('a'))
print(d.match('1'))
```

```python
# single character of alphabet, digit, under bar
d = re.compile(r'\w')

print(d.match(' '))
print(d.match('a'))
print(d.match('1'))
print(d.match('_'))
print(d.match('@'))
print(d.match('ㄱ'))
print(d.match('W'))
```

```python
# special character drain
d = re.compile(r'\W')

print(d.match(' '))
print(d.match('a'))
print(d.match('1'))
print(d.match('_'))
print(d.match('@'))
print(d.match('ㄱ'))
print(d.match('W'))
```
